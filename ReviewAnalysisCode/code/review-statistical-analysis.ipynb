```python
# Script that calculates the distribution of all privacy-related reviews among the categories and outputs it as a pie chart

import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
data = pd.read_excel('filename', sheet_name=None)

# Combine all worksheets into a single DataFrame
combined_df = pd.concat(data.values(), ignore_index=True)

# Define the category columns
category_columns = [
    "policy_related",
    "location_related",
    "data_access_related",
    "permission_related",
    "ad_related",
    "security_related",
    "trust_related",
    "scam_related"
]

# Count the occurrences of "Yes" in the columns of interest
counts = combined_df[category_columns].eq("Yes").sum()

# Calculate the percentages
total_count = counts.sum()
percentages = counts / total_count * 100

# Generate the pie chart
labels = [f"{column} ({percentages[column]:.2f}%)" for column in category_columns]

fig, ax = plt.subplots(figsize=(3.5, 3.5))
ax.pie(counts, startangle=90)
ax.axis("equal")
plt.legend(labels, loc="center left", bbox_to_anchor=(1.1, 0.5))

# Show the pie chart
plt.show()

```

```python
# Script that calculates the distribution of ratings among the respective apps

import pandas as pd

# Read the Excel file
df = pd.read_excel('filename', sheet_name=None)

# Iterate over each worksheet
for worksheet_name, data in df.items():
    # Count the number of reviews for each score
    scores = data['score'].value_counts().sort_index()

    # Calculate the proportion and round to one decimal place
    proportions = round(scores / scores.sum() * 100, 1)

    # Print the percentage distribution for each score
    print(f"Worksheet: {worksheet_name}")
    for score, proportion in proportions.items():
        print(f"Score {score}: {proportion}%")
    print()
```

```python
# Script that calculates the distribution of ratings among the categories

import pandas as pd

# Read the Excel file
df = pd.read_excel('filename', sheet_name=None)

# Combine all worksheets into a single DataFrame
combined_df = pd.concat(df.values(), ignore_index=True)

# Define the category columns
category_columns = [
    "policy_related",
    "location_related",
    "data_access_related",
    "permission_related",
    "ad_related",
    "security_related",
    "trust_related",
    "scam_related"
]

# Calculate and print the percentage distribution for each column with "Yes" values
for column in category_columns:
    # Filter the rows with "Yes" values
    filtered_df = combined_df[combined_df[column] == 'Yes']

    # Count the number of reviews for each score
    scores = filtered_df['score'].value_counts().sort_index()

    # Calculate the proportion and round to one decimal place
    proportions = round(scores / scores.sum() * 100, 1)

    print(f"Column: {column}")
    for score, proportion in proportions.items():
        print(f"Score {score}: {proportion}%")
    print()

```

```python
# Script that calculates the distribution of categories among the ratings

import pandas as pd

# Read the Excel file
data = pd.read_excel('filename', sheet_name=None)

# Combine data from all worksheets
combined_data = pd.concat(data.values(), ignore_index=True)

# Define the category columns
category_columns = [
    "policy_related",
    "location_related",
    "data_access_related",
    "permission_related",
    "ad_related",
    "security_related",
    "trust_related",
    "scam_related"
]

# Calculate the percentage distribution of categories for each score
total_rows = len(combined_data)
scores = combined_data['score'].unique()

for score in scores:
    score_df = combined_data[combined_data['score'] == score]
    score_count = len(score_df)
    print(f"Score {score}:")
    
    category_percentages = []
    
    for column in category_columns:
        category_count = len(score_df[score_df[column] == 'Yes'])
        percentage = (category_count / score_count) * 100
        category_percentages.append(percentage)
    
    # Normalize the percentages
    category_percentages = [percentage / sum(category_percentages) * 100 for percentage in category_percentages]
    
    for column, percentage in zip(category_columns, category_percentages):
        print(f"{column}: {percentage:.1f}%")
    
    print()

```
